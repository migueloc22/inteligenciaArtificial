```python
# Kmeans model

from sklearn.cluster import KMeans
```

```python
# How many clusters (k) are required?
# Elbow method

WCSS = [] #"Within-Cluster-Sum-of-Squares"

for i in range(1,11):
    kmeans = KMeans(n_clusters= i,init='k-means++',random_state=0) #pruebas k
    kmeans.fit(forbes23.iloc[:, 3:]) #entrenamiento
    WCSS.append(kmeans.inertia_) #guardamos resultados
```

```python
print(WCSS)
```

```python
plt.plot(range(1,11),WCSS)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=4, init='k-means++',random_state=0)
```

```python
# Prediction

y = kmeans_model.fit_predict(forbes23.iloc[:,3:])
```
